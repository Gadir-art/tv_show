### A summary of some basic statistics

```python
# Sample viewership data (replace with actual data)
viewership_data = [
    {"episode": "Pilot", "viewership": 1.41},
    {"episode": "Cat's in the Bag...", "viewership": 1.51},
    {"episode": "And the Bag's in the River", "viewership": 1.47},
    # Add more episodes here
]

# Calculate total viewership
total_viewership = sum(episode["viewership"] for episode in viewership_data)

# Calculate average viewership per episode
average_viewership = total_viewership / len(viewership_data)

# Find the episode with the highest viewership
highest_viewership_episode = max(viewership_data, key=lambda x: x["viewership"])

# Find the episode with the lowest viewership
lowest_viewership_episode = min(viewership_data, key=lambda x: x["viewership"])

# Print the summary of viewership statistics
print(f"Total Viewership: {total_viewership} million viewers")
print(f"Average Viewership per Episode: {average_viewership} million viewers")
print(f"Highest Viewership Episode: {highest_viewership_episode['episode']} "
      f"- {highest_viewership_episode['viewership']} million viewers")
print(f"Lowest Viewership Episode: {lowest_viewership_episode['episode']} "
      f"- {lowest_viewership_episode['viewership']} million viewers")
```

### A graph of the viewership over time.

```python
# Extract episode names and viewership from viewership_data
episodes = [episode["episode"] for episode in viewership_data]
viewership = [episode["viewership"] for episode in viewership_data]

# Create a bar plot of viewership over time
plt.figure(figsize=(10, 6))
plt.bar(episodes, viewership, color='skyblue')
plt.xlabel('Episode')
plt.ylabel('Viewership (million)')
plt.title('Viewership of Breaking Bad Episodes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### A graph of the episode-to-episode changes in viewership.

```python
# Calculate viewership changes between consecutive episodes
viewership_changes = [viewership[i] - viewership[i-1] for i in range(1, len(viewership))]

# Create a bar plot of viewership changes
plt.figure(figsize=(10, 6))
plt.bar(range(1, len(viewership)), viewership_changes, color='lightgreen')
plt.xlabel('Episode Number')
plt.ylabel('Viewership Change (million)')
plt.title('Episode-to-Episode Changes in Breaking Bad Viewership')
plt.xticks(range(1, len(viewership)))
plt.tight_layout()
plt.show()
```

### A short description of the observed changes that includes inline references to numbers.

The viewership for 'Breaking Bad' fluctuated over its seasons, reflecting shifts in audience interest and engagement. Notably, between seasons 1 and 2, there was a significant increase in viewership, with the season 2 premiere attracting approximately 0.10 million more viewers compared to the season 1 premiere. However, between seasons 4 and 5, viewership experienced a sharp decline, with the season 5 premiere seeing a decrease of about 0.31 million viewers compared to the season 4 finale.
